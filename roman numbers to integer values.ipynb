```python
#roman to integer converter
''' Symbol       Value
I             1
V             5
X             10
L             50
C             100
D             500
M             1000'''
x=input()
i=0
if "IV" in x:
  x=x.replace("IV" , "IIII")
if "IX" in x:
  x=x.replace("IX","VIIII")
if "XL" in x:
  x=x.replace("XL","XXXX")
if "XC" in x:
  x=x.replace("XC","LXXXX")
if "CD" in x:
  x=x.replace("CD","CCCC")
if "CM" in x:
  x=x.replace("CM","DCCCC")
for Z in x:
  if Z=="I":
    i=i+1
  elif Z=="V":
    i=i+5
  elif Z=="X":
    i=i+10
  elif Z=="L":
    i=i+50
  elif Z=="C":
    i=i+100
  elif Z=="D":
    i=i+500
  elif Z=="M":
    i=i+1000
print(i)
```
